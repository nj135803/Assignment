### Step 8. How many different names exist in the dataset?

```python
unique_names_count = df['Name'].nunique()
print(f"There are {unique_names_count} different name in the dataset.")
```

### Step 9. What is the name with most occurrences?

```python
most_name = df["Name"].value_counts().index
Name_count = df['Name'].value_counts()
print(f"name with most occurrences is\"{most_name[0]}\" with {Name_count[0]} entries")
```

### Step 10. How many different names have the least occurrences?

```python
least_name = df['Name'].value_counts(ascending=True).head().index.tolist()
print(f'The Name(s) with least occurance are: {least_name}')
```

### Step 11. What is the median name occurrence?

```python
name_counts = df['Name'].value_counts()
median_occurrence = name_counts.median()
print(f'The median occurrence of name is {median_occurrence}, ')
```

### Step 12. What is the standard deviation of names?

```python
name_counts = df['Name'].value_counts()
std_deviation = name_counts.std()
print(f'The standard deviation of name occurrence is {std_deviation}, ')
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
name_counts = df['Name'].value_counts()
summary_stats = name_counts.describe()
print(summary_stats)
```
